## Cholesky L Lt

```python
import numpy as np
```

```python
def LU_factorizacion(A):
    n = len(A)
    L = np.eye(n)  # Inicializamos L como una matriz identidad de tamaño n
    U = np.copy(A) # Copiamos A a U para empezar a modificarla en el proceso

    # Proceso de factorización LU
    for i in range(n):
        # Comprobamos que el pivote no sea cero (en este caso asumimos que no lo es, para simplicidad)
        if U[i, i] == 0:
            raise ValueError("La matriz requiere un pivote distinto de cero para factorización LU sin pivoteo.")

        # Iteramos sobre las filas debajo del pivote en la columna i
        for j in range(i + 1, n):
            # Calculamos el factor multiplicador para la fila j
            factor = U[j, i] / U[i, i]
            L[j, i] = factor  # Guardamos el factor en L
            U[j, i:] = U[j, i:] - factor * U[i, i:]  # Actualizamos la fila j en U

    return L, U
```

```python
def Cholesky_LL(L, U):
    n = L.shape[0]

    # Crear una matriz diagonal D a partir de U
    D = np.zeros((n, n), dtype=float)
    for i in range(n):
        D[i, i] = U[i, i]

    L_chol = L @ np.sqrt(D)

    return L_chol, L_chol.T
```

## Cholesky L D Lt

```python
def Cholesky_LDL(L, U):
    n = L.shape[0]

    # Crear una matriz diagonal D a partir de U
    D = np.zeros((n, n), dtype=float)
    for i in range(n):
        D[i, i] = U[i, i]

    return L,D, L.T
```

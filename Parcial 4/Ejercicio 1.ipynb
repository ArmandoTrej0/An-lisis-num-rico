# Bisección

```python
def biseccion(f, a, b, tol, max_iter):
    if f(a) * f(b) >= 0:
        raise ValueError("La función debe cambiar de signo en el intervalo [a, b].")

    contador_iteraciones = 0
    while (b - a) / 2.0 > tol and contador_iteraciones < max_iter:
        c = (a + b) / 2.0  # Punto medio
        if f(c) == 0:
            return c  # Se encontró la raíz exacta
        elif f(a) * f(c) < 0:
            b = c  # La raíz está en el intervalo [a, c]
        else:
            a = c  # La raíz está en el intervalo [c, b]
        contador_iteraciones += 1
    return (a + b) / 2.0
```

# Secante

```python
def secante(f, x0, x1, tol, max_iter):
    contador_iteraciones = 0
    while abs(x1 - x0) > tol and contador_iteraciones < max_iter:
        f_x0 = f(x0)
        f_x1 = f(x1)
        if f_x1 - f_x0 == 0:
            raise ValueError("División por cero en el método de la secante.")
        x2 = x1 - f_x1 * (x1 - x0) / (f_x1 - f_x0)  # Nueva aproximación
        x0, x1 = x1, x2
        contador_iteraciones += 1
    return x1
```

# Newton

```python
def newton(f, df, x0, tol, max_iter):
    contador_iteraciones = 0
    x = x0
    while abs(f(x)) > tol and contador_iteraciones < max_iter:
        x = x - f(x) / df(x)  # Nueva aproximación
        contador_iteraciones += 1
    return x
```

#Regla falsa

```python
def regla_falsa(f, a, b, tol, max_iter):
    if f(a) * f(b) >= 0:
        raise ValueError("La función debe cambiar de signo en el intervalo [a, b].")

    contador_iteraciones = 0
    c = a
    while abs(f(c)) > tol and contador_iteraciones < max_iter:
        c = a - (f(a) * (b - a)) / (f(b) - f(a))  # Intersección de la secante con el eje x
        if f(c) == 0:
            return c  # Se encontró la raíz exacta
        elif f(a) * f(c) < 0:
            b = c  # La raíz está en el intervalo [a, c]
        else:
            a = c  # La raíz está en el intervalo [c, b]
        contador_iteraciones += 1
    return c
```

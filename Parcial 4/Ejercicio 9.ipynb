```python
import math
```

```python
def newton(f, df, x0, tol, max_iter):
    contador_iteraciones = 0
    x = x0
    while abs(f(x)) > tol and contador_iteraciones < max_iter:
        x = x - f(x) / df(x)  # Nueva aproximaciÃ³n
        contador_iteraciones += 1
    return x
```

#a)

```python
def a(x):
  y= math.exp(x) + 2**(-x) + 2*math.cos(x) - 6
  return y
```

```python
def da(x):
  y= math.exp(x) - math.log(2)/2**x - 2*math.sin(x)
  return y
```

```python
y1=newton(a, da, 1, 10**-5, 1000)
y1
```

#b)

```python
def b(x):
  y=2*x*math.cos(2*x) - (x-2)**2
  return y
```

```python
def db(x):
  y=2*math.cos(2*x) - 4*x*math.sin(2*x) - 2*x + 4
  return y
```

```python
y21=newton(b, db, 2, 10**-5, 1000)
y21
```

```python
y22=newton(b, db, 4, 10**-5, 1000)
y22
```

```python
def c(x):
  y= math.exp(x) - 3*(x**2)
  return y
```

```python
def dc(x):
  y= math.exp(x) - 6*x
  return y
```

```python
y31=newton(c, dc, 1, 10**-5, 1000)
y31
```

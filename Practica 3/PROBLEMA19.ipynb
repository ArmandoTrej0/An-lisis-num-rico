```python
import numpy as np

# Matriz A
A = np.array([[0.16, 0.10],
              [0.17, 0.11],
              [2.02, 1.29]])

# Vectores b originales y perturbados
b1 = np.array([0.26, 0.28, 3.31])  # b original (inciso a)
b2 = np.array([0.27, 0.25, 3.33])  # b perturbado (inciso b)

# Función para resolver mínimos cuadrados usando la descomposición QR
def solve_least_squares(A, b):
    Q, R = np.linalg.qr(A)  # Descomposición QR
    return np.linalg.solve(R, Q.T @ b)  # Resolver R * x = Q^T * b

# Resolver para ambos vectores
x1 = solve_least_squares(A, b1)  # Solución para b1
x2 = solve_least_squares(A, b2)  # Solución para b2

# Calcular la diferencia entre las soluciones
delta_x = np.abs(x1 - x2)

# Imprimir resultados
print("Inciso a) Solución para b =", b1)
print("x1 =", x1)
print("\nInciso b) Solución para b perturbado =", b2)
print("x2 =", x2)
print("\nDiferencia entre soluciones |x1 - x2| =")
print(delta_x)

# Análisis del sistema
cond_number = np.linalg.cond(A)
print("\nNúmero de condición de A:", cond_number)

if cond_number > 1e3:
    print("El sistema es mal condicionado (alto número de condición).")
else:
    print("El sistema está bien condicionado (bajo número de condición).")
```

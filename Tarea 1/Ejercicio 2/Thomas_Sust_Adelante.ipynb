```python
import numpy as np

def thomas_adelante(a, b, c, d):
    """Resuelve un sistema de ecuaciones Ax = d usando el algoritmo de Thomas modificado"""
    n = len(b)

    #eliminación hacia adelante
    for i in range(1, n):
        factor = a[i-1] / b[i-1]
        b[i] -= factor * c[i-1]
        d[i] -= factor * d[i-1]

    # sustitución hacia adelante
    x = np.zeros(n)
    x[0] = d[0] / b[0]  # primera solución

    for i in range(1, n):
        x[i] = (d[i] - c[i-1] * x[i-1]) / b[i]

    return x

#Pruebas
a = np.array([1, 1], dtype=float)    # Diagonal inferior
b = np.array([4, 4, 4], dtype=float) # Diagonal principal
c = np.array([1, 1], dtype=float)    # Diagonal superior
d = np.array([5, 5, 5], dtype=float) # Vector de términos independientes

sol = thomas_adelante(a, b, c, d)
print("Solucion:", sol)
```

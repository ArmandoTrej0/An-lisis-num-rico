```python
import numpy as np

def identificar_banda(A):
    """Identifica si la matriz tiene la banda en la diagonal superior o en la diagonal inferior"""
    n = len(A)
    # Revisar si la banda es superior
    for i in range(n - 1):
        if A[i][i+1] != 0:
            return "sup"
    # Si no es superior, asumimos que es inferior
    return "inf"

def resolver(A, b):
    """Resuelve el sistema de ecuaciones lineales"""
    # Identificar si la banda es superior o inferior
    tipo_banda = identificar_banda(A)

    n = len(b)

    if tipo_banda == "sup":
        # Eliminaci칩n para la banda superior
        for i in range(n - 1):
            factor = A[i+1][i] / A[i][i]
            A[i+1][i] = 0  # Eliminamos el coeficiente debajo de la diagonal principal
            A[i+1][i+1] -= factor * A[i][i+1]
            b[i+1] -= factor * b[i]

    elif tipo_banda == "inf":
        # Eliminaci칩n para la banda inferior
        for i in range(1, n):
            factor = A[i][i-1] / A[i-1][i-1]
            A[i][i-1] = 0  # Eliminamos el coeficiente debajo de la diagonal principal
            A[i][i] -= factor * A[i-1][i]
            b[i] -= factor * b[i-1]

    # Sustituci칩n hacia atr치s
    x = np.zeros(n)
    x[-1] = b[-1] / A[-1][-1]

    for i in range(n-2, -1, -1):
        x[i] = (b[i] - A[i][i+1] * x[i+1]) / A[i][i]

    return x

# Ejemplo de uso

# Ejemplo 1: Banda superior
A_superior = np.array([[2, -1, 0], [-1, 2, -1], [0, -1, 2]], dtype=float)
b_superior = np.array([1, 0, 1], dtype=float)
sol_superior = resolver(A_superior, b_superior)
print("La solucion para banda superior:", sol_superior)

# Ejemplo 2: Banda inferior
A_inferior = np.array([[2, 0, 0], [-1, 2, 0], [0, -1, 2]], dtype=float)
b_inferior = np.array([1, 0, 1], dtype=float)
sol_inferior = resolver(A_inferior, b_inferior)
print("La soluciom para banda inferior:", sol_inferior)
```

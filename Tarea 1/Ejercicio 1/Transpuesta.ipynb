```python
A=[[1,2,3,4],[3,4,5,6],[6,7,8,9],[1,4,7,8]]
```

```python
def transpuesta(A):
  #Defino una matriz de 0's del tama√±o requerido (si A es m*n, la transpuesta sera n*m)
  C=[[0 for i in range(len(A))] for j in range(len(A[0]))]

  #Por ultimo solo hago que el elemento j,i de mi nueva matriz sea el elemento i,j de A
  for i in range(len(A)):
    for j in range(len(A[0])):
      C[j][i]=A[i][j]
  return C
```

```python
transpuesta(A)
```

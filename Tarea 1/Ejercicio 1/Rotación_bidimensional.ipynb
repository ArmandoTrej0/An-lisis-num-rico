```python
import math
```

```python
def multiplicar_matrices(A, B):
  if len(A[0]) != len(B):
    return "Las matrices no se pueden multiplicar"
  else:
    C=[[0 for i in range(len(B[0]))] for j in range(len(A))]
    for i in range(len(A)):
      for j in range(len(B[0])):
        for k in range(len(B)):
          C[i][j] += A[i][k] * B[k][j]
    return C
```

```python
def rotación_bidimencional(a,s):  #le pido al usuario un vector en R2 y un angulo
  #Si el vector dado no esta en r2
  if len(a) != 2:
    return "El vector no es bidimensional"

  else:
    #Convierto mi vector en un vector columna
    a=[[a[0]],[a[1]]]

    #Primero hare mi matriz de rotación, la libreria math trabaja en radianes
    s=math.radians(s)
    x=math.cos(s)
    y=math.sin(s)
    R=[[x,-y],[y,x]]

    #Realizo la multiplicación de matrices y regreso mi resultado en formato vector
    C=multiplicar_matrices(R,a)
    return [C[0][0],C[1][0]]
```

```python
q=[1,0]
s=90
rotación_bidimencional(q,s)
#asumimos que 6.123233995736766e-17 es 0 pues es muy pequeño
```

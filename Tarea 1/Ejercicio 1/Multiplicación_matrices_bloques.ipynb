```python
#Adjunto mi codigo anterior pues se usara más adelante
def multiplicar_matrices(A, B):
  if len(A[0]) != len(B):
    return "Las matrices no se pueden multiplicar"
  else:
    C=[[0 for i in range(len(B[0]))] for j in range(len(A))]
    for i in range(len(A)):
      for j in range(len(B[0])):
        for k in range(len(B)):
          C[i][j] += A[i][k] * B[k][j]
    return C
```

```python
#Haremos una función que "una" bloques para hacer una matriz
def unir_submatrices(A1,A2,A3,A4):
  C=[[0 for i in range (len(A1[0])+len(A2[0]))] for j in range(len(A1)+len(A3))]
  for i in range(len(A1)):
    C[i]=A1[i]+A2[i]
  for i in range(len(A3)):
    C[i+len(A1)]=A3[i]+A4[i]
  return C
```

```python
#La función de suma de matrices entrada a entrada
def suma_matrices(A, B):
  if len(A) != len(B) or len(A[0]) != len(B[0]):
    return "Las matrices no son iguales"
  else:
    C=[[A[i][j]+B[i][j] for j in range(len(A[0]))] for i in range(len(A))]
  return C
```

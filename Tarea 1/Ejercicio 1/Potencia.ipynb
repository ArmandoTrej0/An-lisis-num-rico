```python
A=[[1,2,3,4],[3,4,5,6],[6,7,8,9],[1,4,7,8]]
```

```python
def multiplicar_matrices(A, B):
  if len(A[0]) != len(B):
    return "Las matrices no se pueden multiplicar"
  else:
    C=[[0 for i in range(len(B[0]))] for j in range(len(A))]
    for i in range(len(A)):
      for j in range(len(B[0])):
        for k in range(len(B)):
          C[i][j] += A[i][k] * B[k][j]
    return C
```

```python
def potencia_matriz(A,n):
  #Si la matriz no es cuadrada
  if len(A) != len(A[0]):
    return "La matriz no es cuadrada"

  #Por definición una matriz elevada a la 0 es la matriz identidad por lo que si n=0 regreso dicha matriz
  elif n==0:
    return [[1 if i==j else 0 for i in range(len(A))] for j in range(len(A))]

  #De otra manera, usare de manera recursiva la función multiplicar matrices
  else:
    An=A
    for i in range(n-1):
      An=multiplicar_matrices(An,A)
    return An
```

```python
potencia_matriz(A,3)
```

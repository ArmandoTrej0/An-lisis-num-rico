```python
import math
```

```python
def multiplicar_matrices(A, B):
  if len(A[0]) != len(B):
    return "Las matrices no se pueden multiplicar"
  else:
    C=[[0 for i in range(len(B[0]))] for j in range(len(A))]
    for i in range(len(A)):
      for j in range(len(B[0])):
        for k in range(len(B)):
          C[i][j] += A[i][k] * B[k][j]
    return C
```

```python
def rotación_tridimencional(a,s,e): #Pedimos un vector, un angulo y el eje por sobre el cual se rotara
  if e not in ["x","y","z"]:
    return "El eje no es valido"
  elif len(a) != 3:
    return "El vector no es tridimensional"
  else:
    #Convierto mi vector en un vector columna y mi angulo en radianes
    a=[[a[0]],[a[1]],[a[2]]]
    s=math.radians(s)

    #Utilizo las definiciones de matriz de rotacion de cada eje y multiplico
    if e=="x":
      A=[[1,0,0],[0,math.cos(s),-math.sin(s)],[0,math.sin(s),math.cos(s)]]
      C=multiplicar_matrices(A,a)
      return [C[0][0],C[1][0],C[2][0]]
    elif e=="y":
      A=[[math.cos(s),0,math.sin(s)],[0,1,0],[-math.sin(s),0,math.cos(s)]]
      C=multiplicar_matrices(A,a)
      return [C[0][0],C[1][0],C[2][0]]
    else:
      A=[[math.cos(s),-math.sin(s),0],[math.sin(s),math.cos(s),0],[0,0,1]]
      C=multiplicar_matrices(A,a)
      return [C[0][0],C[1][0],C[2][0]]
```

```python
rotación_tridimencional([1,0,0],90,'z')
```

```python
def producto_cruz(a,b):

  #Si los vectores no son tridimensionales
  if len(a) != 3 or len(b) != 3:
    return "Los vectores no son tridimencionales"

  #De otra manera, uso la definici√≥n del producto cruz
  else:
    c1=a[1]*b[2]-a[2]*b[1]
    c2=a[2]*b[0]-a[0]*b[2]
    c3=a[0]*b[1]-a[1]*b[0]
    return [c1,c2,c3]
```

```python
a=[1,0,0]
b=[0,1,0]
producto_cruz(a,b)
```

```python
def norma_1(A):
  #Lista donde anexare las sumas de las columnas
  C=[]

  #Por cada columna sumo todos los elementos de cada fila en esa columna y lo guardo en una lista
  for j in range(len(A[0])):
    s=0
    for i in range(len(A)):
      s=s+abs(A[i][j])
    C.append(s)

  #Regreso el maximo de mi lista de sumas
  return max(C)
```

```python
A=[[-1,2],[7,-8]]
norma_1(A)
```

```python
import math
```

```python
def norma_2(A):
  C=[]
  for i in range(len(A)):
    s=0
    for j in range(len(A[0])):
      s=s+(A[i][j]**2)
      C.append(math.sqrt(s))
  return max(C)
```

```python
A=[[-1,2],[7,-8]]
norma_2(A)
```
